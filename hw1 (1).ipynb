给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9



idea:

````python
# add your code
def question4():
    profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
    n = len(profits)
    if n < 2:
        return 0, -1, -1

    max_profit = float('-inf')
    current_profit = 0
    buy_day = 0
    sell_day = 0
    temp_buy = 0

    for i in range(n):
        current_profit += profits[i]

        if current_profit > max_profit:
            max_profit = current_profit
            buy_day = temp_buy
            sell_day = i

        if current_profit < 0:
            current_profit = 0
            temp_buy = i + 1

    print(max_profit, buy_day + 1, sell_day + 1)
````

idea：对于数组 arr，局部极小值定义为：
在数组起点：arr[0] < arr[1]
在数组终点：arr[n-1] < arr[n-2]
在中间位置 i：满足 arr[i] < arr[i-1] 且 arr[i] < arr[i+1]
代码逐个元素检查是否满足上述条件，并将符合条件的元素加入 local_minima 列表中

````python
# add your code here
def question7(arr):
    n = len(arr)
    if n < 2:
        return arr

    local_minima = []

    for i in range(n):
        if i == 0 and arr[i] < arr[i + 1]:
            local_minima.append(arr[i])
        elif i == n - 1 and arr[i] < arr[i - 1]:
            local_minima.append(arr[i])
        elif 0 < i < n - 1 and arr[i] < arr[i - 1] and arr[i] < arr[i + 1]:
            local_minima.append(arr[i])

    return local_minima
# your algorithm time complexity is :O(n)
````

e.g.：
---

idea：
smk(arr, k)：使用内置 sorted() 获取最小的 k 个数；
smk_heap(arr, k)：用最小堆 heapq 来提取最小的 k 个数。

````python
# add your code here
def question8(arr,k):
    import heapq

    def smk(arr, k):
        sorted_arr = sorted(arr)
        return sorted_arr[:k]

    def smk_heap(arr, k):
        min_heap = []
        for num in arr:
            heapq.heappush(min_heap, num)  

        result = [heapq.heappop(min_heap) for _ in range(k)]  
        return result
# your algorithm time complexity is :O(nlog n)
````

举例1：数组A=["a", "da", "bde", "ab", "bc", "abdc", "cdba"], 排序后的数组应该为：['a', 'ab', 'abdc', 'bc', 'bde', 'cdba', 'da']



举例2：数组A=['ab', 'a', 'b', 'abc', 'ba', 'c'], 排序后的数组应该为：

['a', 'ab', 'abc', 'b', 'ba', 'c']



举例3：数组A=['aef', 'yzr', 'wr', 'ab', 'bhjc', 'lkabdc', 'pwcdba'],  排序后的数组应该为：['ab', 'aef', 'bhjc', 'lkabdc', 'pwcdba', 'wr', 'yzr']



note：

-  两个字符之间的比较可以考虑比较他们对应的ASCII码值；
- python中可以使用`ord("a")`返回字符 “a”对应的ASCII值

idea:

````python
# add your code here
def question10(arr):
    if not arr:
        return arr
    max_len = max(len(s) for s in arr)
    for i in range(max_len - 1, -1, -1):
        buckets = [[] for _ in range(27)]
        for s in arr:
            index = ord(s[i]) - ord('a') + 1 if i < len(s) else 0
            buckets[index].append(s)
        arr = [s for bucket in buckets for s in bucket]

    return arr
# your algorithm time complexity is :O(L⋅n)
````
